```python
import urllib.request
import json
from urllib.error import URLError
import time

# GitHub API-Objekt für Repository-Erkundung
GITHUB_API = "https://api.github.com/repos/louisthecat86/development_suit"

def fetch_github_api(endpoint):
    """Fetch data from GitHub API"""
    url = GITHUB_API + endpoint
    try:
        with urllib.request.urlopen(url) as response:
            return json.loads(response.read().decode())
    except URLError as e:
        return {"error": str(e)}

# Repository-Struktur ermitteln
print("📦 Repository: louisthecat86/development_suit")
print("=" * 60)

# Wurzelverzeichnis erkunden
root_contents = fetch_github_api("/contents/")
print("\n🔍 Verzeichnisstruktur (Wurzelebene):")
if isinstance(root_contents, list):
    for item in sorted(root_contents, key=lambda x: (x['type'] != 'dir', x['name'])):
        icon = "📁" if item['type'] == 'dir' else "📄"
        print(f"  {icon} {item['name']}")
else:
    print(f"  Error: {root_contents}")
```

```python
import urllib.request
import json
import base64
from urllib.error import URLError

GITHUB_API = "https://api.github.com/repos/louisthecat86/Development-Suit"
req = urllib.request.Request(f"{GITHUB_API}/contents/package.json", headers={'User-Agent': 'Mozilla/5.0'})
with urllib.request.urlopen(req) as response:
    data = json.load(response)
package_json = json.loads(base64.b64decode(data['content']).decode())
print('package.json scripts:')
for name, script in package_json.get('scripts', {}).items():
    print(f'  {name}: {script}')
print('\nDependencies with electron or vite:')
for dep, ver in {**package_json.get('dependencies', {}), **package_json.get('devDependencies', {})}.items():
    if 'electron' in dep.lower() or 'vite' in dep.lower() or 'electron-builder' in dep.lower():
        print(f'  {dep}: {ver}')

req = urllib.request.Request(f"{GITHUB_API}/contents/README.md", headers={'User-Agent': 'Mozilla/5.0'})
with urllib.request.urlopen(req) as response:
    data = json.load(response)
readme = base64.b64decode(data['content']).decode()
print('\nREADME first lines:')
print('\n'.join(readme.splitlines()[:20]))
```
